```python
import numpy as np
import pandas as pd
```

```python
#Load the data into a Pandas DataFrame and display the first five rows
sales=pd.read_csv('sales_data_sample.csv', encoding='latin1')
sales
```

```python
#To Get first five rows
sales.head(5)
```

```python
#Display the column names and their data types.
sales.columns
```

```python
sales.dtypes
```

```python
#Count the total number of unique ORDERNUMBER values
unique_order_count = sales['ORDERNUMBER'].nunique()
unique_order_count
```

```python
#Filter and display only the orders where the STATUS is "Shipped" and PRODUCTLINE is"Motorcycles."
sales_status = sales[(sales['STATUS'] == 'Shipped') & (sales['PRODUCTLINE'] == 'Motorcycles')]
sales_status
```

```python
#Find all orders with a SALES value greater than 4000 and display their ORDERNUMBER,SALES, and ORDERDATE.
higher_sales_orders = sales[sales['SALES'] > 4000]
new_df=higher_sales_orders[['ORDERNUMBER', 'SALES', 'ORDERDATE']]
new_df
```

```python
#Display orders from the year 2004 with a QUANTITYORDERED greater than 30.
sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'])
filtered_orders=sales[(sales['ORDERDATE'].dt.year == 2004) & (sales['QUANTITYORDERED'] > 30)]
filtered_orders
```

```python
#Calculate the total SALES for each YEAR_ID and PRODUCTLINE.
years = sales['YEAR_ID'].unique()
products = sales['PRODUCTLINE'].unique()
total_sales = [
    {
        'YEAR_ID': year,
        'PRODUCTLINE': product,
        'TOTAL_SALES': sales[(sales['YEAR_ID'] == year) & (sales['PRODUCTLINE'] == product)]['SALES'].sum()
    }
    for year in years
    for product in products
] 
total_df = pd.DataFrame(total_sales)
total_df
```

```python
#Find the average PRICEEACH for each unique PRODUCTLINE
productlines = sales['PRODUCTLINE'].unique()
avg_prices = [
    {
        'PRODUCTLINE': product,
        'AVG_PRICEEACH': sales[sales['PRODUCTLINE'] == product]['PRICEEACH'].mean()
    }
    for product in productlines
]
avg_price_df = pd.DataFrame(avg_prices)
avg_price_df
```

```python
#Identify the month (MONTH_ID) with the highest SALES in 2003.
sales_of_2003 = sales[sales['YEAR_ID'] == 2003]
monthly_sales_2003 = sales_of_2003.groupby('MONTH_ID')['SALES'].sum()
highest_month = monthly_sales_2003.idxmax()
highest_sales = monthly_sales_2003.max()
highest_month,highest_sales
```

```python
#Sort the DataFrame by ORDERDATE in ascending order and display the first five rows.
sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'])
updated_sales = sales.sort_values(by='ORDERDATE', ascending=True)
updated_sales.head(5)
```

```python
#Find the top 3 orders with the highest SALES and display their ORDERNUMBER, SALES, andPRODUCTLINE.
top_sales = sales.sort_values(by='SALES', ascending=False)
top_3_orders = top_sales[['ORDERNUMBER', 'SALES', 'PRODUCTLINE']].head(3)
top_3_orders
```

```python
#Create a new column called TOTAL_PRICE that calculates the total price for each order (QUANTITYORDERED * PRICEEACH).
sales['TOTAL_PRICE'] = sales['QUANTITYORDERED'] * sales['PRICEEACH']
```

```python
#Convert the ORDERDATE column to a datetime format and extract the month name, saving it in a new column called MONTH_NAME.
sales['ORDERDATE'] = pd.to_datetime(sales['ORDERDATE'])
sales['MONTH_NAME'] = sales['ORDERDATE'].dt.strftime('%B')
sales[['ORDERDATE','MONTH_NAME']]
```

```python
#Replace any PRODUCTLINE labeled "Classic Cars" with "Vintage Cars"
sales['PRODUCTLINE'] = sales['PRODUCTLINE'].replace('Classic Cars', 'Vintage Cars')
```

```python
#Group the data by YEAR_ID and calculate the total SALES for each year.
yearly_sales = sales.groupby('YEAR_ID')['SALES'].sum().reset_index()
yearly_sales
```

```python
#Use a pivot table to display the total SALES for each PRODUCTLINE by QTR_ID.
pivot_table_sales = sales.pivot_table(
    values='SALES',
    index='PRODUCTLINE',
    columns='QTR_ID',
    aggfunc='sum'
)
pivot_table_sales
```

```python
#Find the MSRP average for each PRODUCTLINE across all orders.
avg_msrp = sales.groupby('PRODUCTLINE')['MSRP'].mean().reset_index()
avg_msrp
```

```python
#What is the total quantity ordered for each PRODUCTLINE?
total_quantity = sales.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().reset_index()
total_quantity
```

```python
#How many orders were placed each quarter (QTR_ID) in 2003?
sales_in_2003 = sales[sales['YEAR_ID'] == 2003]
orders_in_quarter_2003 = sales_in_2003.groupby('QTR_ID')['ORDERNUMBER'].nunique().reset_index()
orders_in_quarter_2003
```

```python
#Which ORDERNUMBER had the highest SALES in 2005?
sales_in_2005 = sales[sales['YEAR_ID'] == 2005]
order_sales_in_2005 = sales_in_2005.groupby('ORDERNUMBER')['SALES'].sum()
highest_order = order_sales_in_2005.idxmax()
highest_sales = order_sales_in_2005.max()
highest_order,highest_sales
```

```python
sales.to_csv('sales_data.csv',index=False)
```
